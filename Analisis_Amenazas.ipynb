```python
data_2014 = pd.read_csv('/home/dalev/Escritorio/Datathon/data/Amenazas/amenazas_2014_0(Document_CH2138).csv', delimiter = ';', encoding = 'latin1')
data_2014
```

```python
data_2014.head(15)
```

```python
#Copia del dataframe original
data2014 = data_2014.copy()

#Eliminamos las columnas que no son de interés
data2014 = data2014.drop([0, 1, 2, 3, 4, 5, 6, 7, 8])

#Reiniciamos los índices y ponemos las columnas como la primera fila
data2014 = data2014.reset_index(drop=True)
data2014.columns = data2014.iloc[0]
data2014 = data2014.drop([0])
data2014 = data2014.reset_index(drop=True)

#Mostramos el nuevo dataframe
data2014.head(15)
```

```python
#Revision de los valores nulos en la data
data2014.isnull().sum()
```

```python
#Revisión de los datos de la columna "DEPARTAMENTO" cuyo valor sea igual a "ANTIOQUIA"
data_antioquia_2014 = data2014[data2014['DEPARTAMENTO'] == 'ANTIOQUIA']
```

```python
data_antioquia_2014['MUNICIPIO'].unique()
```

```python
#Revisión de los datos de la columna "DEPARTAMENTO" cuyo valor sea igual a "ANTIOQUIA" y "MUNICIPIO" sea igual a "CALDAS", "LA ESTRELLA", "ITAGUI", "ENVIGADO", "SABANETA",  "MEDELLÍN (CT)", "BELLO", "COPACABANA", "GIRARDOTA" Y "BARBOSA"
data_2014_areaMetropolitana = data_antioquia_2014[data_antioquia_2014['MUNICIPIO'].isin(['CALDAS', 'LA ESTRELLA', 'ITAGUI', 'ENVIGADO', 'SABANETA',  'MEDELLÍN (CT)', 'BELLO', 'COPACABANA', 'GIRARDOTA', 'BARBOSA'])]
```

```python
data_2014_areaMetropolitana.info()
```

```python
#Cambio del tipo de dato de la columna "FECHA HECHO" a datetime
data_2014_areaMetropolitana['FECHA HECHO'] = pd.to_datetime(data_2014_areaMetropolitana['FECHA HECHO'], format='%d/%m/%Y')

#Separación de la columna "FECHA HECHO" en "AÑO", "MES" y "DÍA"
data_2014_areaMetropolitana['AÑO'] = data_2014_areaMetropolitana['FECHA HECHO'].dt.year
data_2014_areaMetropolitana['MES'] = data_2014_areaMetropolitana['FECHA HECHO'].dt.month
data_2014_areaMetropolitana['DÍA'] = data_2014_areaMetropolitana['FECHA HECHO'].dt.day

#Eliminación de la columna "FECHA HECHO"
data_2014_areaMetropolitana = data_2014_areaMetropolitana.drop(columns=['FECHA HECHO'])

#Mostramos el nuevo dataframe
data_2014_areaMetropolitana.head(10)
```

```python
#Reduzco el dataframe a las columnas "MUNICIPIO", "AÑO", "MES", "DÍA" Y "CANTIDAD"
timeserie_2014_areaMetropolitana = data_2014_areaMetropolitana[['MUNICIPIO', 'AÑO', 'MES', 'DÍA', 'CANTIDAD']]
```

```python
#Mostramos el nuevo dataframe
timeserie_2014_areaMetropolitana.head(10)
```

```python
timeserie_2014_areaMetropolitana.shape
```

```python
#Agrupo la variable municio con las variables de mes y dia, con el fin de obtener el número de eventos por municipio en cada mes y día
data_2014_areaMetropolitana_grouped = data_2014_areaMetropolitana.groupby(['MUNICIPIO', 'MES', 'DÍA']).size().reset_index(name='NÚMERO DE EVENTOS')

#Mostramos el nuevo dataframe
data_2014_areaMetropolitana_grouped.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ordenar el DataFrame por 'AÑO' y 'MUNICIPIO'
time_Series_Amenazas = time_Series_Amenazas.sort_values(by=['AÑO', 'MUNICIPIO'])

# Crear el gráfico de serie de tiempo
plt.figure(figsize=(20, 10))
sns.lineplot(data=time_Series_Amenazas, x='AÑO', y='CANTIDAD', hue='MUNICIPIO')
plt.title('Cantidad de amenazas por año en cada municipio')
plt.xlabel('Año')
plt.ylabel('Cantidad de amenazas')
plt.legend(title='Municipio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Invertir el eje y (no recomendado para gráficos de línea)
plt.gca().invert_yaxis()

plt.show()
```
